#### Linear Search

```python
class LinearSearch:
    def linearSearch(self, arr, ele):
        for i in range(len(arr)):
            if(arr[i]==ele):
                return "Element Found at index : ", i
        return "Element Not Found"
obj = LinearSearch()
obj.linearSearch([5,1,9,3,8,0,4,2,7],2)
```

#### Binary Search

```python
class BinarySearch:
    def binarySearch(self, arr, ele):
        left = 0
        right = len(arr)-1
        while(left<=right):
            mid = (left+right)//2
            print(mid, end = '\t')
            if(arr[mid]==ele):
                return "Element Found at index : ", mid
            elif(arr[mid]>ele):
                right = mid - 1
            else:
                left = mid + 1
        return "Element Not Found"
obj = BinarySearch()
arr = [5,1,9,3,8,0,4,2,7,6]
arr.sort()
print("sorted array ::: ", arr)
obj.binarySearch(arr,0)
```
