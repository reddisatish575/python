```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        
    def constructTree(self, arr ,root, i, n):
        if i<n:
            root = TreeNode(arr[i-1])
            root.left = self.constructTree(arr, root.left, 2*i, n)
            root.right = self.constructTree(arr, root.right, 2*i+1, n)
        return root
        
    def preOrder(self, root, result):
        if not root: return []
        result.append(root.value)
        self.preOrder(root.left, result)
        self.preOrder(root.right, result)
        return result
        
    def inOrder(self, root, result):
        if not root: return []
        self.inOrder(root.left, result)
        result.append(root.value)
        self.inOrder(root.right, result)
        return result
        
    def postOrder(self, root, result):
        if not root : return []
        self.postOrder(root.left, result)
        self.postOrder(root.right, result)
        result.append(root.value)
        return result
```

```python
arr = [7,2,8,0,1,3,4,6,9,5]
root = None
obj = TreeNode(root)
binaryTree = obj.constructTree(arr, root ,1 , len(arr)+1)

print(binaryTree)
```

```python
print("The Tree Elements :: ", arr)
print("Pre Order Result of the Tree is :: ", obj.preOrder(binaryTree, []))
print("In Order Result of the Tree is :: ", obj.inOrder(binaryTree, []))
print("Post Order Result of the Tree is :: ", obj.postOrder(binaryTree, []))
```

```python
class TreeNode:
    def __init__(self,val=None):
        self.val=val
        self.left=None
        self.right=None

    def constructTree(self,arr,root,i,n):
        if i<n:
            root=TreeNode(arr[i])
            root.left=self.constructTree(arr,root.left,2*i+1,n)
            root.right=self.constructTree(arr,root.right,2*i+2,n)
        return root
    def preOrder(self,root,res=[]):
        if not root:
            return []
        res.append(root.val)
        self.preOrder(root.left,res)
        self.preOrder(root.right,res)
        return res
        
    def inOrder(self,root,res=[]):
        if not root: return []
        self.inOrder(root.left,res)
        res.append(root.val)
        self.inOrder(root.right,res)
        return res
    
    def postOrder(self,root,res=[]):
        if not root:
            return []
        self.postOrder(root.left,res)
        self.postOrder(root.right,res)
        res.append(root.val)
        return res

    def leafNodes(self,root,res=[]):
        if not root.left and not root.right:
            res.append(root.val)
        if root.left: self.leafNodes(root.left,res)
        if root.right: self.leafNodes(root.right,res)
        return res
arr=[1,2,3,4,5,6,7]
obj=TreeNode()
root=obj.constructTree(arr,None,0,len(arr))
print(root)
```
