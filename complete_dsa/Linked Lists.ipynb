```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        print("coming to display")
        if not self.head:
            return "Lined List is Empty"
        lstr = ''
        itr = self.head
        while itr:
            lstr+=str(itr.data)+'-->' if itr.next else str(itr.data)
            itr = itr.next
        return lstr

    def insertAtBeg(self, val):
        node = Node(val)
        if not self.head:
            self.head = node
        # temp = self.head
        node.next = self.head
        self.head = node
        return self.display()

    def inserList(self, arr):
        print("coming to inserList ::")
        itr = self.head
        if not itr:
            self.head = Node(arr[0])
            itr = self.head
            for i in arr[1:]:
                itr.next = Node(i)
                itr = itr.next
        else:
            while itr.next:
                itr=itr.next
            for i in arr:
                itr.next = Node(i)
                itr = itr.next
        return self.display()
obj = LinkedList()
arr = [5,3,9,1,2,0,8,7,6,4]
obj.inserList(arr)
```
