```python
class MergeSort:
    
    def divide(self, arr):
        if(len(arr)<=1):
            return arr
        mid = len(arr)//2
        left = self.divide(arr[:mid])
        right = self.divide(arr[mid:])
        return self.conquer(left, right)

    def conquer(self, left, right):
        result = []
        i,j = 0, 0
        while(i<len(left) and j<len(right)):
            if(left[i]<right[j]):
                result+=[left[i]]
                i+=1
            else:
                result+=[right[j]]
                j+=1
        result+=left[i:]+right[j:]
        return result

obj = MergeSort()
arr = [9,2,6,1,5,8,0,4,7,-2,54,-32,4,-2321]
obj.divide(arr)
```

```python
class QuickSort:
    def quickSort(self, arr, left, right):
        if left<right:
            partitionPosition = self.partition(arr, left, right)
            self.quickSort(arr, left, partitionPosition-1)
            self.quickSort(arr, partitionPosition+1, right)
        return arr

    def partition(self, arr, left, right):
        i = left
        j = right-1
        pivot = arr[right]
        while i<j:
            while i<right and arr[i]<pivot:
                i+=1
            while j>=i and arr[j]>pivot:
                j-=1
            if i<j:
                arr[i], arr[j] = arr[j], arr[i]
        if(arr[i]>pivot):
            arr[i],arr[right] = arr[right], arr[i]
        return i
obj = QuickSort()
arr = [9,2,6,1,5,8,0,4,7,-2,54,-32,4,-2321]
obj.quickSort(arr, 0, len(arr)-1)
```

```python
class QuickSort:
   def quickSort(self, arr):
       if(len(arr)<=1):
           return arr
       else:
           pivot = arr[0]
           left = [i for i in arr[1:] if i < pivot]
           right = [i for i in arr[1:] if i>=pivot]
           return self.quickSort(left) + [pivot] + self.quickSort(right)
           
obj = QuickSort()
arr = [9,2,6,1,5,8,0,4,7,-2,54,-32,4,-2321]
obj.quickSort(arr)
```

```python
class HeapSort:

    def buildHeapTree(self, arr, n):
        for i in range(n//2-1,-1,-1):
            self.buildHeap(arr, n, i)
        self.heapSort(arr, n)

    def buildHeap(self, arr, n, i):
        largest = i
        left = 2*i+1
        right = 2*i+2

        if(left<n and arr[left]>arr[largest]):
            largest = left
        if(right<n and arr[right]> arr[largest]):
            largest = right
        if(largest!=i):
            arr[largest], arr[i] = arr[i], arr[largest]
            self.buildHeap(arr, n, largest)

    def printHeapTree(self, arr):
        for i in arr:
            print(i, end=', ')

    def heapSort(self, arr, n):
        for i in range(n-1, -1, -1):
            arr[i], arr[0] = arr[0], arr[i]
            self.buildHeap(arr, i, 0)
        self.printHeapTree(arr)

obj = HeapSort()
arr = [9,2,6,1,5,8,0,4,7,-2,54,-32,4,-2321]
obj. buildHeapTree(arr, len(arr))
```

```python
class HeapSort:

    def heapify(self, arr, n, i):
        largest = i
        left = 2*i+1
        right = 2*i+2

        if left < n and arr[left] > arr[largest]:
            largest = left
        if right < n and arr[right] > arr[largest]:
            largest = right
        if largest != i:
            arr[i], arr[largest] = arr[largest], arr[i]
            self.heapify(arr, n, largest)

    def heapSort(self, arr):
        n = len(arr)
        # Step 1: Build max heap
        for i in range(n//2 - 1, -1, -1):
            self.heapify(arr, n, i)

        # Step 2: Extract elements one by one
        for i in range(n-1, 0, -1):
            arr[0], arr[i] = arr[i], arr[0]
            self.heapify(arr, i, 0)

        return arr

obj = HeapSort()
arr = [9,2,6,1,5,8,0,4,7,-2,54,-32,4,-2321]
print(obj.heapSort(arr))
```
