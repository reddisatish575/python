#### Base Conversions
1. Decimal base 10 (No prefix)
2. Binary base 2 (0b or 0B as prefix)
3. Octal base 8 (0o or 0O as prefix)
4. Hexa base 16 (0x or 0X as prefix)

```python
print(bin(15))
print(bin(0o11))
print(bin(0x10))
print(oct(10))
print(oct(0B1111))
print(oct(0X123))
print(hex(100))
print(hex(0B11111))
print(hex(0o12345))
```

#### Immutability
1. All Fundamental datatypes (int, float, complex, bool,str) are immutable
2. Once we creates an object,we cannot perform any changes in that object.
3. If we are trying to change then with those changes a new object will be created.

#### How Object is Created ?
1. In Python if a new object is required, then PVM wont create object immediately. First it will check is any object available with the required content or not.
2. If available then existing object will be reused. If it is not available then only a new object will be created.
3. The advantage of this approach is memory utilization and performance will be improved.

#### How Address to the variable is assigned ?
1. For fundamental data type, address is assigned based on the Value not the variable name.
2. For Non-Fundamental data type, every variable address is different.

```python
a= 10+5j
b= 10+5j
print(id(a),id(b)) # not fundamental datatype, so addreess changed.
print(a==b)   # it compares the value assigned to the address
print(a is b) # it compares address of the variable
```

```python
a= 10
b=10
print(id(a),id(b)) # fundamental datatype, so addreess is same.
print(a==b)
print(a is b)
```

#### Bytes data type
1. bytes data type represens a group of byte numbers just like an array.
2. The only allowed values for byte data type are 0 to 256. By mistake if we are trying to provide any other values then we will get value error.
3. Once we creates bytes data type value, we cannot change its values,otherwise we will get TypeError.

```python
x= [10,20,30,40]
b= bytes(x)
print(b)
print(b[2])
```

```python
b[0] = 100
```

```python
x = [0,100,256] 
print(bytes(x))
```

#### Bytearray data type
1. bytearray is exactly same as bytes data type except that its elements can be modified. 

```python
x= [10,20,30,40]
b= bytearray(x)
print(b)
```

```python
b[0]=100
print(b[0])
```

```python
x = [0,100,256] 
print(bytearray(x))
```

#### Range data type
1. range Data Type represents a sequence of numbers.
2. The elements present in range Data type are not modifiable. i.e range Data type is
immutable.
3. Syntax : range(start, end, step)
4. We can access elements present in the range Data Type by using index.
5. We cannot modify the values of range data type
6. We can create a list of values with range data type

```python
a= range(10)
for i in a: print(i, end=', ')
```

```python
a[0]=100
```

```python
a= range(10)
print(list(a))
```

#### Frozenset data type
1. It is exactly same as set except that it is immutable.
2. Hence we cannot use add or remove functions.

```python
s= {1,2,3,4}
fs= frozenset(s)
print(fs)
```
